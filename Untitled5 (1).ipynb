```python
import pandas as pd
import numpy as np
import torch
import torch.nn.functional as F
from torch_geometric.data import Data
from torch_geometric.nn import SAGEConv
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import NearestNeighbors
import networkx as nx
import matplotlib.pyplot as plt

# Sample DataFrame (Replace with your actual data)
data = {
    'Latitude': np.random.uniform(12, 15, 100),
    'Longitude': np.random.uniform(75, 80, 100),
    'Gender': np.random.choice(['Male', 'Female'], 100),
    'time_section': np.random.choice(['Morning', 'Afternoon', 'Evening', 'Night'], 100),
    'accident_score': np.random.randint(1, 10, 100)
}

# Load your dataset
df = pd.DataFrame(data)

# Select 30 unique locations based on latitude and longitude
unique_locations = df[['Latitude', 'Longitude']].drop_duplicates().head(30)
selected_indices = unique_locations.index
df_selected = df.loc[selected_indices]

# Encode categorical variables
categorical_features = ['Gender', 'time_section']
le_dict = {}
for feature in categorical_features:
    le = LabelEncoder()
    df_selected[feature] = le.fit_transform(df_selected[feature])
    le_dict[feature] = le

# Select relevant features
features = ['Latitude', 'Longitude', 'Gender', 'time_section']
X = df_selected[features]

# Standardize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define target variable
y = df_selected['accident_score']

# Construct graph with nearest neighbors
def find_nearest_neighbors(data, num_neighbors=5):
    nbrs = NearestNeighbors(n_neighbors=num_neighbors + 1, algorithm='ball_tree').fit(data)
    distances, indices = nbrs.kneighbors(data)
    return indices[:, 1:], distances[:, 1:]  # Exclude the point itself

def construct_graph_with_neighbors(features, target, num_neighbors=5):
    neighbor_indices, distances = find_nearest_neighbors(features, num_neighbors)

    edge_index = []
    edge_attr = []
    for node_idx, (neighbors, dist) in enumerate(zip(neighbor_indices, distances)):
        for neighbor_idx, distance in zip(neighbors, dist):
            edge_index.append([node_idx, neighbor_idx])
            edge_index.append([neighbor_idx, node_idx])  # Ensure undirected edges
            edge_attr.append([distance])
            edge_attr.append([distance])

    edge_index = torch.tensor(edge_index, dtype=torch.long).t().contiguous()
    edge_attr = torch.tensor(edge_attr, dtype=torch.float)
    x = torch.tensor(features, dtype=torch.float)
    y = torch.tensor(target, dtype=torch.float).unsqueeze(1)  # Ensure target is 2D for regression

    return Data(x=x, edge_index=edge_index, edge_attr=edge_attr, y=y)

# Create the graph data for the selected locations
graph_data = construct_graph_with_neighbors(X_scaled, y, num_neighbors=5)

# Define the GraphSAGE Model
class GraphSAGEModel(torch.nn.Module):
    def __init__(self, num_node_features, hidden_dim, output_dim):
        super(GraphSAGEModel, self).__init__()
        self.conv1 = SAGEConv(num_node_features, hidden_dim)
        self.conv2 = SAGEConv(hidden_dim, output_dim)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        return x

# Initialize the GraphSAGE model
model = GraphSAGEModel(num_node_features=X_scaled.shape[1], hidden_dim=16, output_dim=1)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

# Train the model
def train(model, train_data, num_epochs=100):
    model.train()
    for epoch in range(num_epochs):
        optimizer.zero_grad()
        output = model(train_data)
        loss = F.mse_loss(output, train_data.y)
        loss.backward()
        optimizer.step()
        if epoch % 10 == 0:
            print(f'Epoch {epoch}, Loss: {loss.item():.4f}')

# Train the model on the graph data
train(model, graph_data)

# Add a new location to the graph
def add_new_location_to_graph(existing_graph, new_node_features, num_neighbors=5):
    new_node_features = torch.tensor(new_node_features, dtype=torch.float).reshape(1, -1)
    combined_features = np.vstack((existing_graph.x.numpy(), new_node_features))

    # Find nearest neighbors including the new node
    neighbor_indices, distances = find_nearest_neighbors(combined_features, num_neighbors)

    # Create new edge indices including the new node
    new_edge_index = []
    new_edge_attr = []
    num_existing_nodes = existing_graph.num_nodes

    for node_idx, (neighbors, dist) in enumerate(zip(neighbor_indices, distances)):
        for neighbor_idx, distance in zip(neighbors, dist):
            new_edge_index.append([node_idx, neighbor_idx])
            new_edge_index.append([neighbor_idx, node_idx])  # Ensure undirected edges
            new_edge_attr.append([distance])
            new_edge_attr.append([distance])

    new_edge_index = torch.tensor(new_edge_index, dtype=torch.long).t().contiguous()
    new_edge_attr = torch.tensor(new_edge_attr, dtype=torch.float)
    new_x = torch.cat([existing_graph.x, new_node_features], dim=0)

    return Data(x=new_x, edge_index=new_edge_index, edge_attr=new_edge_attr)

# Example new location data (latitude, longitude, gender, time_section)
new_location = {
    'Latitude': 13.0827,
    'Longitude': 80.2707,
    'Gender': 'Male',
    'time_section': 'Morning'
}

# Convert new location data to DataFrame for consistency with preprocessing
new_location_df = pd.DataFrame([new_location])

# Encode and scale the features using the same transformations
for feature in categorical_features:
    new_location_df[feature] = le_dict[feature].transform(new_location_df[feature])

new_location_features = new_location_df[features]
new_location_scaled = scaler.transform(new_location_features)

# Create a new graph with the additional node
new_graph_data = add_new_location_to_graph(graph_data, new_location_scaled, num_neighbors=5)

# Function to update edge accident scores based on user input
def update_accident_scores(graph_data, source_node, target_node, new_accident_score):
    edge_index = graph_data.edge_index.t().numpy()
    edge_attr = graph_data.edge_attr.numpy()

    for i, (src, tgt) in enumerate(edge_index):
        if (src == source_node and tgt == target_node) or (src == target_node and tgt == source_node):
            edge_attr[i] += new_accident_score  # Update with new accident score
            break

    graph_data.edge_attr = torch.tensor(edge_attr, dtype=torch.float)
    return graph_data

# Example of updating accident score between nodes (user input)
source_node = int(input("Enter source node: "))
target_node = int(input("Enter target node: "))
new_accident_score = float(input("Enter new accident score: "))

# Update the graph with the new accident data
updated_graph_data = update_accident_scores(new_graph_data, source_node, target_node, new_accident_score)

# Function to visualize the entire graph
def visualize_graph(graph_data):
    G = nx.Graph()
    for i, (src, tgt) in enumerate(graph_data.edge_index.t().numpy()):
        G.add_edge(src, tgt, weight=graph_data.edge_attr[i].item())

    pos = nx.spring_layout(G, seed=42)  # Positions for all nodes

    plt.figure(figsize=(10, 7))

    # Draw the graph
    nx.draw_networkx_nodes(G, pos, node_size=500, node_color='lightblue')
    nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.5)

    # Draw node labels
    nx.draw_networkx_labels(G, pos, font_size=12, font_color='black')

    plt.title("Graph Visualization")
    plt.show()

# Visualize the entire graph
visualize_graph(updated_graph_data)

# Make predictions using the trained model
model.eval()
with torch.no_grad():
    output = model(updated_graph_data)
    # The last node's output is the prediction for the new location
    predicted_danger_score = output[-1].item()

print(f"Predicted Danger Score for the new location: {predicted_danger_score:.4f}")
```

```python
import pandas as pd
import torch
from torch_geometric.data import Data
from torch_geometric.nn import SAGEConv
import torch.nn.functional as F
import torch.nn as nn
import networkx as nx

# Sample data (replace this with your actual data)
nodes_data = {
    'node_id': [0, 1, 2, 3, 4, 5],
    'latitude': [10, 20, 30, 40, 50, 60],
    'longitude': [15, 25, 35, 45, 55, 65],
    'accident_score': [1, 2, 0.5, 3, 2.5, 1]
}
edges_data = {
    'source': [0, 1, 2, 3, 4],
    'target': [1, 2, 3, 4, 5],
    'distance': [10, 20, 30, 40, 50],
    'accident_score': [0, 0, 0, 0, 0]  # Initial edge accident scores
}

nodes_df = pd.DataFrame(nodes_data)
edges_df = pd.DataFrame(edges_data)

# Create node features tensor (e.g., accident score)
node_features = torch.tensor(nodes_df[['accident_score']].values, dtype=torch.float)

# Create edge index tensor
edge_index = torch.tensor(edges_df[['source', 'target']].values.T, dtype=torch.long)

# Create edge attribute tensor (e.g., distance and accident score)
edge_attr = torch.tensor(edges_df[['distance', 'accident_score']].values, dtype=torch.float)

# Create graph data object
data = Data(x=node_features, edge_index=edge_index, edge_attr=edge_attr)

# Define the GraphSAGE model
class SafeRouteGraphSAGE(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(SafeRouteGraphSAGE, self).__init__()
        self.conv1 = SAGEConv(input_dim, hidden_dim)
        self.conv2 = SAGEConv(hidden_dim, output_dim)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = F.relu(self.conv1(x, edge_index))
        x = self.conv2(x, edge_index)
        return x

# Initialize the model
model = SafeRouteGraphSAGE(input_dim=1, hidden_dim=16, output_dim=1)

# Training setup
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = nn.MSELoss()  # Adjust this based on your task

# Training loop
model.train()
for epoch in range(100):  # Number of epochs
    optimizer.zero_grad()
    out = model(data)
    loss = criterion(out, data.x)  # Adjust this based on your task
    loss.backward()
    optimizer.step()
    print(f'Epoch {epoch+1}, Loss: {loss.item()}')

# Convert to NetworkX graph
G = nx.Graph()
for i, row in edges_df.iterrows():
    G.add_edge(row['source'], row['target'], weight=row['distance'])

# Function to update the graph with new accident data on an edge
def update_graph_with_new_accident(edge_accidents_df, data, model):
    # Update edge features with new accident scores
    for i, row in edge_accidents_df.iterrows():
        source = row['source']
        target = row['target']
        new_score = row['accident_score']

        # Update the edge attribute tensor in data
        edge_index = (data.edge_index[0] == source) & (data.edge_index[1] == target)
        edge_index_rev = (data.edge_index[0] == target) & (data.edge_index[1] == source)  # To cover undirected edge case
        data.edge_attr[edge_index, 1] = torch.tensor([new_score], dtype=torch.float)
        data.edge_attr[edge_index_rev, 1] = torch.tensor([new_score], dtype=torch.float)

        # Recalculate node scores using the trained model
        with torch.no_grad():
            model.eval()
            node_scores = model(data).squeeze().numpy()

        # Update edge weights based on new edge scores and node scores
        G[source][target]['weight'] = edges_df[(edges_df['source'] == source) & (edges_df['target'] == target)]['distance'].values[0] + new_score + node_scores[source] + node_scores[target]
        G[target][source]['weight'] = G[source][target]['weight']  # To cover undirected edge case

# Example of updating the graph with new accident data on an edge
edge_accidents_df = pd.DataFrame({'source': [1], 'target': [2], 'accident_score': [5]})
update_graph_with_new_accident(edge_accidents_df, data, model)

# Find the safest path using Dijkstra's algorithm
start_node = 0  # Example start node
end_node = 5    # Example end node
safest_path = nx.dijkstra_path(G, source=start_node, target=end_node, weight='weight')
print('Safest Path:', safest_path)
```
